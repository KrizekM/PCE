```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = "M.xlsx"  
df = pd.read_excel(file_path)

sm_cols = [col for col in df.columns if str(col).startswith("SM")]

# průměr a směrodatná odchylka (převod z mm na m)
meanr = df[sm_cols].mean().values / 100000
standartr = df[sm_cols].std().values / 100000

meanr_raw = df[sm_cols].mean().values
standartr_raw = df[sm_cols].std().values

print("Průměr (meanr) – před dělením:\n", meanr_raw)
print("\nSměrodatná odchylka (standartr) – před dělením:\n", standartr_raw)

# Ořez na délku nodes 
if len(meanr) > len(nodes):
    meanr = meanr[:len(nodes)]
    standartr = standartr[:len(nodes)]
elif len(meanr) < len(nodes):
    raise ValueError("Počet SM sloupců menší než počet uzlů.")

Fmultiplier = meanr
std_Fmultiplier = standartr

# Výpočet radiálních deformací od středu
center = nodes.mean(axis=0)

def deform_nodes(nodes, delta):
    out = np.zeros_like(nodes)
    for i in range(len(nodes)):
        v = nodes[i] - center
        norm = np.linalg.norm(v)
        if norm != 0:
            out[i] = nodes[i] + delta[i] * (v / norm)
        else:
            out[i] = nodes[i]
    return out

# 3σ
axialF_mean   = deform_nodes(nodes, Fmultiplier)
axialF_plus3  = deform_nodes(nodes, Fmultiplier + 3*std_Fmultiplier)
axialF_minus3 = deform_nodes(nodes, Fmultiplier - 3*std_Fmultiplier)

plt.figure(figsize=(16.5 / 2.54, 5.2))

# Tunel a deformace (otevřené křivky)
plt.plot(nodes[:, 0],        nodes[:, 1],        'b-',  label='Tunel (původní)')
plt.plot(axialF_mean[:, 0],  axialF_mean[:, 1],  'r-',  label='Střední deformace')
plt.plot(axialF_plus3[:, 0], axialF_plus3[:, 1], 'g--', label='+3σ deformace')
plt.plot(axialF_minus3[:, 0], axialF_minus3[:, 1], 'g--', label='−3σ deformace')

# Spojení horního bodu
plt.plot([nodes[-1, 0],        nodes[0, 0]],        [nodes[-1, 1],        nodes[0, 1]],        'b-')
plt.plot([axialF_mean[-1, 0],  axialF_mean[0, 0]],  [axialF_mean[-1, 1],  axialF_mean[0, 1]],  'r-')
plt.plot([axialF_plus3[-1, 0], axialF_plus3[0, 0]], [axialF_plus3[-1, 1], axialF_plus3[0, 1]], 'g--')
plt.plot([axialF_minus3[-1, 0], axialF_minus3[0, 0]], [axialF_minus3[-1, 1], axialF_minus3[0, 1]], 'g--')

plt.xlabel('X [m]')
plt.ylabel('Y [m]')
plt.title('Deformace tunelu M DATA')
plt.grid(True)
plt.axis('equal')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
