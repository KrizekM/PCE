```python
csv_path = Path("Oakwood3_excel_ready.csv")  
SURFACE_SCALE = 150                       # zvětšení povrchových posunů


df = pd.read_csv(csv_path)
center = nodes.mean(axis=0)

#  tunel
A_points = {
    251:(0.0,-10.5275), 260:(-4.29059,-11.4844), 281:(-7.17776,-13.5229),
    289:(-9.49344,-17.4055), 295:(-9.64331,-20.3488), 403:(-9.27246,-21.7842)
}
A_cols_x = {uid: f"U_{uid}_1" for uid in A_points}
A_cols_y = {uid: f"U_{uid}_2" for uid in A_points}

#  povrch zeminy 
B_points = {
    320:(-28.0,0.0), 318:(-20.0,0.0), 316:(-12.0,0.0), 315:(-8.0,0.0),
    314:(-4.0,0.0), 313:(0.0,0.0), 312:(4.0,0.0), 311:(8.0,0.0),
    310:(12.0,0.0), 308:(20.0,0.0), 306:(28.0,0.0)
}
B_cols = {uid: f"U_{uid}_19.2" for uid in B_points}

def stats_two(colx, coly, frame):
    "Vektorový průměr (ux, uy), mean(|u|), 3σ(|u|), N."
    if colx not in frame.columns or coly not in frame.columns:
        return np.nan, np.nan, np.array([np.nan, np.nan]), 0
    ux = pd.to_numeric(frame[colx], errors="coerce").to_numpy()
    uy = pd.to_numeric(frame[coly], errors="coerce").to_numpy()
    total = np.sqrt(ux**2 + uy**2)
    dd = 1 if np.sum(~np.isnan(total)) > 1 else 0
    mean_mag = float(np.nanmean(total))
    three_sigma = 3.0 * float(np.nanstd(total, ddof=dd))
    mean_vec = np.array([np.nanmean(ux), np.nanmean(uy)])
    N = int(np.sum(~np.isnan(ux) & ~np.isnan(uy)))
    return mean_mag, three_sigma, mean_vec, N

def stats_one(col, frame):
    "Skalární průměr (např. svislý posun na povrchu), 3σ, N."
    if col not in frame.columns:
        return np.nan, np.nan, 0
    v = pd.to_numeric(frame[col], errors="coerce").to_numpy()
    dd = 1 if np.sum(~np.isnan(v)) > 1 else 0
    return float(np.nanmean(v)), 3.0 * float(np.nanstd(v, ddof=dd)), int(np.sum(~np.isnan(v)))


A_stats = {uid: stats_two(A_cols_x[uid], A_cols_y[uid], df) for uid in A_points}
B_stats = {uid: stats_one(B_cols[uid], df) for uid in B_points if B_cols[uid] in df.columns}

plt.figure(figsize=(11, 7))

# Obrys tunelu
plt.plot(nodes[:, 0], nodes[:, 1], label="Tunel (původní)")
plt.plot([nodes[-1, 0], nodes[0, 0]], [nodes[-1, 1], nodes[0, 1]])  # uzavření

end_means_A = {}
for uid, (x, y) in A_points.items():
    mean_mag, three_sigma, mean_vec, _N = A_stats[uid]
    start = np.array([x, y])
    end_mean = start + mean_vec
    end_means_A[uid] = end_mean.copy()

    d = end_mean - start
    n = np.linalg.norm(d)
    if n < 1e-12:
        v = start - center
        d = v / (np.linalg.norm(v) + 1e-12)
    else:
        d = d / n

    plt.arrow(start[0], start[1], *(end_mean - start),
              length_includes_head=True, head_width=0.15, head_length=0.3)
    seg0 = end_mean - three_sigma * d
    seg1 = end_mean + three_sigma * d
    plt.plot([seg0[0], seg1[0]], [seg0[1], seg1[1]], linewidth=1)
    plt.annotate(str(uid), (x, y), xytext=(6, 6), textcoords="offset points")

xs, ys, labels, sigmas = [], [], [], []
for uid, (x, y) in B_points.items():
    if uid not in B_stats:
        continue
    mean, three_sigma, _N = B_stats[uid]
    xs.append(x)
    ys.append(y + mean * SURFACE_SCALE)            
    labels.append(uid)
    sigmas.append(three_sigma)

if xs:
    plt.plot(xs, ys, '-o', label=f"Povrchové body (mean) ×{int(SURFACE_SCALE)}")
    for i, (x, y, uid, s) in enumerate(zip(xs, ys, labels, sigmas)):
        plt.annotate(str(uid), (x, y), xytext=(0, -14), textcoords="offset points", ha="center")
        if uid in (318, 308):
            offset_y = 12
        else:
            offset_y = 12 if i % 2 == 0 else 22
        plt.annotate(
            f"3σ:{s*1000:.2f}",
            (x, y),
            xytext=(0, offset_y),
            textcoords="offset points",
            ha="center",
            fontsize=9,
            bbox=dict(facecolor='white', edgecolor='none', boxstyle='round,pad=0.2', alpha=1.0)
        )

# abulka
rows = []
for uid, st in A_stats.items():
    rows.append({
        "ID": uid,
        "mean |u| [m]": st[0],
        "3σ [mm]": st[1]*1000,
        "mean ux [m]": st[2][0],
        "mean uy [m]": st[2][1],
        "Skupina": "Tunel"
    })
for uid, st in B_stats.items():
    rows.append({
        "ID": uid,
        "mean |u| [m]": abs(st[0]),
        "3σ [mm]": st[1]*1000,
        "mean ux [m]": 0,
        "mean uy [m]": st[0],
        "Skupina": "Povrch"
    })
summary_df = pd.DataFrame(rows).sort_values(["Skupina", "ID"]).reset_index(drop=True)

try:
    from IPython.display import display
    display(summary_df)
except Exception:
    print(summary_df.to_string(index=False))

#Zvýraznění maxima
if not summary_df.empty and summary_df["mean |u| [m]"].notna().any():
    idx_max = summary_df["mean |u| [m]"].idxmax()
    max_row = summary_df.loc[idx_max]
    max_id = int(max_row["ID"])
    max_val_mm = float(max_row["mean |u| [m]"]) * 1000.0  

    px = py = None
    if max_id in A_points:
        px, py = end_means_A[max_id]
    elif max_id in B_points:
        base_x, base_y = B_points[max_id]
        mean_val = float(max_row["mean uy [m]"])   # skutečná hodnota v metrech
        px, py = base_x, base_y + mean_val * SURFACE_SCALE

    if px is not None:
        plt.plot(px, py, 'o', color='red', markersize=10, zorder=5, label="Max posun")
        plt.annotate(
            f"{max_val_mm:.3f} mm",
            (px, py),
            xytext=(0, -30),
            textcoords="offset points",
            color="red",
            fontsize=12,
            fontweight="bold",
            ha="center"
        )

# zhled grafu
plt.gca().set_aspect("equal", adjustable="box")
plt.grid(True)
plt.xlabel("X [m]")
plt.ylabel("Y [m])".format(SURFACE_SCALE))
plt.title("Tunel: DATA".format(SURFACE_SCALE))
plt.legend(loc="lower left")
plt.tight_layout()
plt.show()
```
